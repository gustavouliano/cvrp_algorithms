```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
cenarios = ["P-n16-k8", "P-n19-k2", "P-n20-k2", "P-n21-k2", "E-n22-k4", "E-n23-k3", "P-n23-k8", "E-n30-k3", 
            "E-n33-k4", "E-n51-k5", "E-n76-k7", "E-n76-k8"]

num_clientes = [16, 19, 20, 21, 22, 23, 23, 30, 33, 51, 76, 76]

melhor_sa = [462, 222, 221, 220, 397, 658, 547, 558, 884, 669, 1063, 1021]
melhor_gb = [456, 212, 217, 212, 381, 640, 554, 544, 893, 712, 1292, 1212]
melhor_ga = [451, 221, 218, 216, 381, 624, 531, 550, 928, 874, 1373, 1447]

media_sa = [510, 247, 252, 240, 454, 704, 592, 669, 972, 755, 1142, 1188]
media_gb = [464, 215, 218, 217, 389, 725, 572, 579, 964, 803, 1453, 1382]
media_ga = [451, 227, 230, 225, 402, 686, 535, 625, 987, 905, 1491, 1516]

tempo_sa = [120, 99, 103, 106, 121, 123, 162, 145, 158, 226, 319, 325]
tempo_gb = [1340, 1375, 1429, 1604, 1656, 1907, 2070, 2217, 2340, 3284, 4749, 4455]
tempo_ga = [9743, 6195, 6216, 6311, 8000, 9450, 11822, 10700, 11613, 14329, 18761, 19749]

x = np.array(num_clientes)
width = 2
```

```python
df = pd.DataFrame({
    "Número de Clientes": num_clientes * 3,
    "Média da Solução": media_sa + media_gb + media_ga,
    "Meta-heurística": (["Simulated Annealing"] * len(num_clientes)) +
                       (["Golden Ball"] * len(num_clientes)) +
                       (["GA Modelo Ilhas"] * len(num_clientes))
})
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=df,
    x="Número de Clientes",
    y="Média da Solução",
    hue="Meta-heurística",
    marker="o"
)
plt.title("Número de clientes X Valor médio da solução", fontsize=14)
plt.xlabel("Número de Clientes", fontsize=12)
plt.ylabel("Média da olução", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend(title="Meta-heurística")
plt.tight_layout()
plt.show()
```

```python
largura_barra = 0.25
posicoes = np.arange(len(num_clientes))

plt.figure(figsize=(12, 6))

plt.bar(posicoes - largura_barra, media_sa, width=largura_barra, label="Simulated Annealing")
plt.bar(posicoes, media_gb, width=largura_barra, label="Golden Ball")
plt.bar(posicoes + largura_barra, media_ga, width=largura_barra, label="GA Modelo Ilhas")

plt.xticks(posicoes, num_clientes)
plt.xlabel("Número de clientes", fontsize=12)
plt.ylabel("Média da solução", fontsize=12)
plt.legend(title="Meta-heurística")
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
data = {
    'Número de Clientes': num_clientes,
    'Simulated Annealing': tempo_sa,
    'Golden Ball': tempo_gb,
    'GA Modelo Ilhas': tempo_ga
}
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(12, 6))
for algoritmo in ['Simulated Annealing', 'Golden Ball', 'GA Modelo Ilhas']:
    sns.lineplot(
        data=df, x='Número de Clientes', y=algoritmo, marker='o', label=algoritmo
    )

plt.xlabel('Número de clientes', fontsize=12)
plt.ylabel('Tempo médio (ms)', fontsize=12)
plt.legend(title="Meta-heurística")
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
