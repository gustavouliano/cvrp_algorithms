```python
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
import tsplib95
import itertools as it
import random
import vrplib
import math
import pandas as pd
import time
from operator import attrgetter
```

```python
def draw_graph(G):
    pos = {i: (data['x'], data['y']) for i, data in G.nodes(data=True)}
    plt.figure(figsize=(8, 6))    
    nx.draw_networkx_nodes(G, pos, node_size=300, node_color='lightblue')
    nx.draw_networkx_edges(G, pos)
    labels = {i: f'{i} (d={data["demand"]})' for i, data in G.nodes(data=True)}
    nx.draw_networkx_labels(G, pos, labels, font_size=10)
    plt.title("Gr√°fico de nodos VRP (Vehicle Routing Problem)")
    plt.xlabel("Coordenada X")
    plt.ylabel("Coordenada Y")
    plt.grid(True)
    plt.show()
```

```python
G = nx.MultiDiGraph()

instance = vrplib.read_instance("./Vrp-Set-P/P/P-n23-k8.vrp")
demand = instance['demand']
nodes_coords = instance['node_coord']
for i, (x, y) in enumerate(nodes_coords):
    G.add_node(i, x=int(x), y=int(y), demand=int(demand[i]))

G.edges(data=True)
draw_graph(G)
```

```python
def euclidean_distance(node1, node2):
    x1, y1 = G.nodes[node1]['x'], G.nodes[node1]['y']
    x2, y2 = G.nodes[node2]['x'], G.nodes[node2]['y']
    return math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
```

```python
def inicializar_solucao(G, capacidade):
    deposito = 0
    nos = list(G.nodes)
    nos.remove(deposito)
    random.shuffle(nos)
    
    rotas = []
    rota_atual = [deposito]
    capacidade_atual = 0

    for no in nos:
        demanda = G.nodes[no]['demand']
        if capacidade_atual + demanda > capacidade:
            rota_atual.append(deposito)
            rotas.append(rota_atual)
            rota_atual = [deposito]
            capacidade_atual = 0
        rota_atual.append(no)
        capacidade_atual += demanda

    rota_atual.append(deposito)
    rotas.append(rota_atual)

    return rotas


def calcular_custo_rota(rota):
    custo = 0
    for i in range(len(rota) - 1):
        custo += euclidean_distance(rota[i], rota[i + 1])
    return custo


def calcular_custo_total(solucao):
    return sum(calcular_custo_rota(rota) for rota in solucao)

def swap_move(solucao, G, capacidade):
    nova_solucao = [rota[:] for rota in solucao]
    rota1, rota2 = random.sample(nova_solucao, 2)
    
    if len(rota1) > 2 and len(rota2) > 2:
        idx1 = random.randint(1, len(rota1) - 2)
        idx2 = random.randint(1, len(rota2) - 2)
        demanda1 = G.nodes[rota1[idx1]]['demand']
        demanda2 = G.nodes[rota2[idx2]]['demand']
        capacidade_rota1 = sum(G.nodes[no]['demand'] for no in rota1 if no != 0)
        capacidade_rota2 = sum(G.nodes[no]['demand'] for no in rota2 if no != 0)
        
        if (capacidade_rota1 - demanda1 + demanda2 <= capacidade and
            capacidade_rota2 - demanda2 + demanda1 <= capacidade):
            rota1[idx1], rota2[idx2] = rota2[idx2], rota1[idx1]
    
    return nova_solucao


def shift_move(solucao, G):
    nova_solucao = [rota[:] for rota in solucao]
    rota = random.choice(nova_solucao)
    if len(rota) > 3:
        idx1, idx2 = random.sample(range(1, len(rota) - 1), 2)
        rota[idx1], rota[idx2] = rota[idx2], rota[idx1]
    
    return nova_solucao


def gerar_vizinho(solucao, G, capacidade):
    if random.random() < 0.5:
        return swap_move(solucao, G, capacidade)
    else:
        return shift_move(solucao, G)



def simulated_annealing(G, capacidade, temperatura_inicial, taxa_resfriamento, max_iteracoes):
    solucao_atual = inicializar_solucao(G, capacidade)
    melhor_solucao = solucao_atual
    custo_atual = calcular_custo_total(solucao_atual)
    melhor_custo = custo_atual
    temperatura = temperatura_inicial

    for iteracao in range(max_iteracoes):
        vizinho = gerar_vizinho(solucao_atual, G, capacidade)
        custo_vizinho = calcular_custo_total(vizinho)

        if custo_vizinho < custo_atual or random.random() < math.exp((custo_atual - custo_vizinho) / temperatura):
            solucao_atual = vizinho
            custo_atual = custo_vizinho

            if custo_atual < melhor_custo:
                melhor_solucao = solucao_atual
                melhor_custo = custo_atual

        temperatura *= taxa_resfriamento

    return melhor_solucao, melhor_custo
```

```python
# Par√¢metros
capacidade = instance['capacity']
temperatura_inicial = 5000
taxa_resfriamento = 0.999 
iteracoes = 10000

melhor_solucao, melhor_custo = simulated_annealing(G, capacidade, temperatura_inicial, taxa_resfriamento, iteracoes)

print("Melhor solu√ß√£o: ", melhor_solucao)
print("Custo da melhor solu√ß√£o: ", melhor_custo)

print(capacidade)
i = 1
for rota in melhor_solucao:
    capacidade_rota = 0
    for nodo in rota:
        capacidade_rota += G.nodes[nodo]['demand']
    print(f'Capacidade {i}: {capacidade_rota}')
    i += 1
```

```python
def testar_parametros(combinacoes_parametros, G, capacidade, num_testes):
    resultados = []
    
    for temp_inicial, taxa_resf, max_iter in combinacoes_parametros:
        custos = []
        tempos = [] 
        
        for _ in range(num_testes):
            inicio = time.time()
            melhor_solucao, melhor_custo = simulated_annealing(G, capacidade, temp_inicial, taxa_resf, max_iter)
            fim = time.time()
            
            custos.append(melhor_custo)
            tempos.append((fim - inicio) * 1000)
        
        media_custos = sum(custos) / num_testes
        melhor_custo_global = min(custos)
        media_tempo = sum(tempos) / num_testes
        
        resultados.append({
            'Temperatura inicial': temp_inicial,
            'Taxa de resfriamento': taxa_resf,
            'M√°ximo de itera√ß√µes': max_iter,
            'M√©dia dos custos': media_custos,
            'Melhor custo': melhor_custo_global,
            'M√©dia do tempo (ms)': media_tempo
        })
        
        print(f"Par√¢metros: T={temp_inicial}, Resf={taxa_resf}, Iter={max_iter} -> M√©dia Custo: {media_custos:.2f}, Melhor Custo: {melhor_custo_global:.2f}, M√©dia Tempo: {media_tempo:.2f} ms")
    
    return resultados
```

```python
temperaturas_iniciais = [16000]
taxas_resfriamento = [0.998]
max_iteracoes_list = [8000]
capacidade = instance['capacity']

combinacoes_parametros = list(it.product(temperaturas_iniciais, taxas_resfriamento, max_iteracoes_list))
```

```python
resultados = testar_parametros(combinacoes_parametros, G, capacidade, 15)
df_resultados = pd.DataFrame(resultados)
# df_resultados.to_csv('resultados_simulated_annealing.csv', index=False)

print(df_resultados.sort_values(by='M√©dia dos custos').head(10))
```

```python
def desenhar_rotas(G, solucao):
    posicoes = {i: (data['x'], data['y']) for i, data in G.nodes(data=True)}
    plt.figure(figsize=(8, 6))
    
    nx.draw_networkx_nodes(G, posicoes, node_size=300, node_color='lightblue')
    nx.draw_networkx_labels(G, posicoes, font_size=10)
    
    for rota in solucao:
        arestas = [(rota[i], rota[i + 1]) for i in range(len(rota) - 1)]
        nx.draw_networkx_edges(G, posicoes, edgelist=arestas, edge_color='r', alpha=0.6)
    
    plt.title("Rotas")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.show()

desenhar_rotas(G, melhor_solucao)
```
